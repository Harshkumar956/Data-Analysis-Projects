## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Dropping columns --> company, agent owing to possessing huge missing values

df.drop(columns =['company', 'agent'], inplace=True) # inplace = True to make permanent changes in the dataframe
````

````python
# Converting column reservation_status_date into datetime dtype using pandas' method to_datetime...

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
````

````python
df['reservation_status_date'].info()
# Now it is successfully converted into datetime64[ns]/datetime...
````

````python
# Dropping missing values... As there are barely 488 missing values,
# therefore, dropping 488 rows will barely have any impact on our dataset

df.dropna(subset=['country', 'children'], inplace=True) # inplace = True to make permanent changes in the dataset
````

````python
# Removing the outliers i.e. removing the rows where adr = 5400 and adr >0 i.e. there are no negative values

df = df[(df['adr'] != 5400) & (df['adr']>=0)] # creating a new dataframe which will possess the rows where adr is not equal to 5400 and there are no negative values i.e. adr greater than 0
````

````python
# Checking if outliers are gone...

df.describe()

# There are no outliers in the dataset now
````

````python
# Creaing a new column by the name of month based on reservation_status_date

df['month'] = df['reservation_status_date'].dt.month_name()

df['month'].head()
````

````python
# Checking if our dataset still has missing values...

df.isnull().sum()

# We have successfully removed all the missing values
# Now we have a ready dataset to perform our analysis...
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code

# Group data by hotel type and calculate total bookings and cancellations
total_bookings = df.groupby('hotel')['is_canceled'].count() # total no. of bookings
canceled_bookings = df.groupby('hotel')['is_canceled'].sum() # total no. of canceled bookings as value is 1 so we added the values to get the count.

try:
  # Plot the first pie chart for City Hotel by creating a blank figure of size 7, 4 units
  plt.figure(figsize=(7, 4))

  # Now, plotting subplots(when plotting more than 1 graphs in a single blank figure) of 2 pie charts... 1,2,1 is the dimension i.e. 1 rows, 2 columns, 1st plot
  plt.subplot(1, 2, 1)

  # plotting pie chart providing first argument for not canceled bookings and second for canceled in city hotel
  # eventually getting a pie chart showing percentage of canceled and not canceled bookings in city hotel
  plt.pie([total_bookings['City Hotel'] - canceled_bookings['City Hotel'], canceled_bookings['City Hotel']],
          labels=['Not Canceled', 'Canceled'], autopct='%1.1f%%', startangle=90)
  plt.title('City Hotel') # providing the title to the first plot

  # Plot the second pie chart for Resort Hotel
  plt.subplot(1, 2, 2)
  plt.pie([total_bookings['Resort Hotel'] - canceled_bookings['Resort Hotel'], canceled_bookings['Resort Hotel']],
          labels=['Not Canceled', 'Canceled'], autopct='%1.1f%%', startangle=90)
  plt.title('Resort Hotel') # providing the title to the second plot


  plt.suptitle('Percentage of Cancellations by Hotel Type') # Setting the title for the whole plot
  plt.axis('equal') # to align both the plots
  plt.show()
except Exception as e:
  print(e)
````

##### 1. Why did you pick the specific chart?

I have picked a pie chart to get the information about the cancellations percentage in both City and Resort Hotels... because the pie chart provides the best and comprehensive visualization of any data in terms of percentage.

##### 2. What is/are the insight(s) found from the chart?

Here we can find that approx. **42%** of the bookings in City Hotel are cancelled whereas, **28%** of bookings are cancelled in Resort Hotels...



##### 3. Will the gained insights help creating a positive business impact?

````python
# Chart - 5 visualization code

try:
  # Group data by month and cancellation status
  monthly_cancellations = df.groupby(['arrival_date_month', 'is_canceled']).size().unstack()

  # Calculate cancellation rate for each month
  cancellation_rate = monthly_cancellations[1] / (monthly_cancellations[0] + monthly_cancellations[1]) * 100

  # Plot the line plot for cancellation rates
  plt.plot(cancellation_rate.index, cancellation_rate.values, marker='o')

  plt.title('Seasonal Trends in Cancellation Rates')
  plt.xlabel('Month')
  plt.ylabel('Cancellation Rate (%)')
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()

except Exception as e:
  print(e)
````

#### Chart - 8

````python
# Chart - 8 visualization code

try:
  #Extract month from 'reservation_status_date'
  df['reservation_status_month'] = df['reservation_status_date'].dt.to_period('M')

  # Filter the DataFrame for canceled and not canceled bookings
  canceled_data = df[df['is_canceled'] == 1]
  not_canceled_data = df[df['is_canceled'] == 0]

  # Calculate mean cancellation rate for different ADR values over time (e.g., months)
  canceled_adr_cancellation_rates = canceled_data.groupby('reservation_status_month')['adr'].mean()
  not_canceled_adr_cancellation_rates = not_canceled_data.groupby('reservation_status_month')['adr'].mean()

  # Plot the lines
  plt.figure(figsize = (14, 4))
  plt.plot(canceled_adr_cancellation_rates.index.astype(str), canceled_adr_cancellation_rates.values, marker='o', label='Canceled', color='red')
  plt.plot(not_canceled_adr_cancellation_rates.index.astype(str), not_canceled_adr_cancellation_rates.values, marker='o', label='Not Canceled', color='blue')

  plt.title('Impact of ADR on Cancellations Over Time')
  plt.xlabel('Month')
  plt.ylabel('Average Daily Rate (ADR)')
  plt.legend()
  plt.xticks(rotation=45)
  plt.grid(True)
  plt.show()

except Exception as e:
  print(e)
````
